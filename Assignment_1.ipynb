## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
df = pd.DataFrame() 
df['time'] = pd.date_range('1/1/2023', periods = 41) 
print(df['time'])
df['year'] = df['time'].dt.year 
df['month'] = df['time'].dt.month 
df['day'] = df['time'].dt.day
```
